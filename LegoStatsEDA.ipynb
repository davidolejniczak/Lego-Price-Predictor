Lego Set EDA notebook, looking at trends in lego set data and pricing 

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('data/LegoStatsHistorical.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['num_parts'], df['retail_price'], alpha=0.4, color='red')

plt.title('Piece Count vs. Retail Price of LEGO Sets', fontsize=16)
plt.xlabel('Piece Count', fontsize=12)
plt.ylabel('Retail Price (CAD)', fontsize=12)
plt.grid(axis='both', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
